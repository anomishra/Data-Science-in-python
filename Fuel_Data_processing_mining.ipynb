## Data Cleaning: Processing the missing data
Drop or replace are two options.
Replace by frequency or replace by mean or any other function. Drop the whole row or column.

```
#Calculate the average of column
mean_col_normalized_losses = data["normalized-losses"].astype("float").mean(axis = 0)
mean_col_normalized_losses
```

```
# Replace the NaN by mean value in "normalized-losses" column
data["normalized-losses"].replace(np.nan, mean_col_normalized_losses, inplace = True)
data["normalized-losses"]
```

Repeat the same calculations for 'bore', 'stroke', 'horsepower', 'peak-rpm'.

```
mean_bore = data['bore'].astype('float').mean(axis=0)
data['bore'].replace(np.nan, mean_bore, inplace= True)

mean_bore = data['stroke'].astype('float').mean(axis=0)
data['stroke'].replace(np.nan, mean_bore, inplace= True)

mean_horsepower = data['horsepower'].astype('float').mean(axis=0)
data['horsepower'].replace(np.nan, mean_horsepower, inplace= True)

mean_peak_rpm = data['peak-rpm'].astype('float').mean(axis=0)
data['peak-rpm'].replace(np.nan, mean_peak_rpm, inplace= True)
```

Another column is 'num-of-doors'. The data element is four or two. Hence check the statistics of the data set.
And replace the missing values by the element having maximum frequency.

```
# check the frequency of all the data elements in the column.
data['num-of-doors'].value_counts()

#                      OR

# Directly see the maximum frequency element.
data['num-of-doors'].value_counts().idxmax()
print("Maximum occurred element is: ",data['num-of-doors'].value_counts().idxmax())

#replace by four, four is having the maximum frequency.
data["num-of-doors"].replace(np.nan, "four", inplace = True)
data["num-of-doors"]
```

Now, see the dataset again. Check the 'price' column and the number of null values inside it.
Since it is very small, hence drop all the values from the rows having empty dataset. Instead you can follow other methods described above. Just for practice, dropping the entrire row. Also reset the index afterwards.


```
data['price'].isnull().sum()
#print('The number of missing values are: ',data['price'].isnull().sum())
print("\n")
# Drop the row with Nan value.
data.dropna(subset=["price"], axis=0, inplace = True)
data['price']

# Reset the index
data.reset_index(drop = True, inplace = True)

# Check the dataset
data.head()
```

The data column 'normalized-losses', 'price', 'peak-rpm' ,'bore' and 'stroke' variables are numerical values, but it is showing as object type. It should be in float or int type format.

## Data Standarization

This is the process to transform the data into meaningful comparisons.
Here we will transform the mpg to L/100 km interms of fuel consumption in the dataset. The formula for unit conversion is L/100km = 235 / mpg.


```
# transform mpg to L/100km 
data['city-L/100km'] = 235/data["city-mpg"]
#data['city-L/100km']

#similarly 
data['highway-L/100km'] = 235/data["highway-mpg"]
#data['highway-L/100km']


#Renaming the column names
data.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)
data.rename(columns={'"city-mpg"':'city-L/100km'}, inplace=True)
```

## Data Normalisation
Normalization is the process of transforming values of several variables into a similar range. Scaling includes the variable average is 0, scaling the variable so the variable variance is 1, or scaling variable so the variable values range from 0 to 1. Here,  I'm normalizing "length", "width" and "height" in the range of [0,1].

```
# replace (origianl value) by (original value)/(maximum value)
data['length'] = data['length']/data['length'].max()
data['width'] = data['width']/data['width'].max()
#data['length']
data['height'] = data['height']/data['height'].max()
```

## Binning
Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.
 "horsepower" is a real valued variable ranging from 48 to 288, with different values. Tranforming the elements into high, medium and low as per the prices.

```
data["horsepower"]=data["horsepower"].astype(float, copy=True)

#bin_size = 4
binwidth = (max(data["horsepower"])-min(data["horsepower"]))/4

bins = np.arange(min(data["horsepower"]), max(data["horsepower"]), binwidth)
print(bins)

#group name
group_names = ['Low', 'Medium', 'High']

data['horsepower-binned'] = pd.cut(data['horsepower'], bins, labels=group_names,include_lowest=True )
data[['horsepower','horsepower-binned']].head(20)
```

##One hot encoding

When there is a categorical data and the data is unable to get inferences, then one hot enoding can be precessed to make the algorithms to adapt and predict. In pandas, it can be done via providing dummies.

```
dummy_variable_1 = pd.get_dummies(data["fuel-type"])
print(dummy_variable_1.head())

#changeing the name of the columns
dummy_variable_1.rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
dummy_variable_1.head()

```

Above the diesel is 0 and the gas is 1.
Now merging the one hot encoding into the original data.

```
# merge data frame "data" and "dummy_variable_1" 
data = pd.concat([data, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "data"
data.drop("fuel-type", axis = 1, inplace=True)
```
