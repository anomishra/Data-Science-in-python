This is the process to transform the data into meaningful comparisons.

```
# transform mpg to L/100km 
data['city-L/100km'] = 235/data["city-mpg"]
#data['city-L/100km']

#similarly 
data['highway-L/100km'] = 235/data["highway-mpg"]
#data['highway-L/100km']


#Renaming the column names
data.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)
data.rename(columns={'"city-mpg"':'city-L/100km'}, inplace=True)
```

## Data Normalisation
Normalization is the process of transforming values of several variables into a similar range. Scaling includes the variable average is 0, scaling the variable so the variable variance is 1, or scaling variable so the variable values range from 0 to 1. Here,  I'm normalizing "length", "width" and "height" in the range of [0,1].

```
# replace (origianl value) by (original value)/(maximum value)
data['length'] = data['length']/data['length'].max()
data['width'] = data['width']/data['width'].max()
#data['length']
data['height'] = data['height']/data['height'].max()
```

## Binning
Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.
 "horsepower" is a real valued variable ranging from 48 to 288, with different values. Tranforming the elements into high, medium and low as per the prices.

```
data["horsepower"]=data["horsepower"].astype(float, copy=True)

#bin_size = 4
binwidth = (max(data["horsepower"])-min(data["horsepower"]))/4

bins = np.arange(min(data["horsepower"]), max(data["horsepower"]), binwidth)
print(bins)

#group name
group_names = ['Low', 'Medium', 'High']

data['horsepower-binned'] = pd.cut(data['horsepower'], bins, labels=group_names,include_lowest=True )
data[['horsepower','horsepower-binned']].head(20)
```

##Correlation

Correlation tells how the variables are related or depending on each other.
1. Positive Correlation: both variables change in the same direction.
2. Neutral Correlation: No relationship in the change of the variables.
3. Negative Correlation: variables change in opposite directions.

Explanation:
1. Positive correlation means that they depending each other in same direction. Ex:  6 seater SUV have higher price from 4 seater (if they belongs to same manufacturing company***), i.e., number of seats increasing means price is increasing.
2. Negative correlation means, again the data is depending but in opposite direction, i.e., if one variable is decreasig then other is increasing.
3. Neutral correlation means, no relationship between variables.

Types of correlation is :
1. Covariance
2. Pearson's correlation- returns a value between -1 and 1. Notable correlation is - return is above 0.5 or below -0.5. 0 means no relation.
3. Spearman's correlation

## Correlation and casuation

 **Correlation**: a measure of the extent of interdependence between variables.

**Causation**: the relationship between cause and effect between two variables.

It is important to know the difference between these two and that correlation does not imply causation. Determining  correlation is much simpler  the determining causation as causation may require independent experimentation.



Pearson correlation is described above.

**P-value**: 
What is this P-value? The P-value is the probability value that the correlation between these two variables is statistically significant. Normally, we choose a significance level of 0.05, which means that we are 95% confident that the correlation between the variables is significant. 

By convention, when the 
- p-value is < 0.001 we say there is strong evidence that the correlation is significant,
- the p-value is < 0.05; there is moderate evidence that the correlation is significant,
- the p-value is < 0.1; there is weak evidence that the correlation is significant, and
- the p-value is > 0.1; there is no evidence that the correlation is significant.

##Anova (Analysis of variables)

A statistical method to test the statistically significant differences between the means of three or more independent (unrelated) groups. Here, the groups are within a one variable. For example, drive-wheels have three groups, i.e., 4wd, fwd, and rwd . There are two parameters in ANOVA to be measured: 

**F-test score**: ANOVA assumes the means of all groups are the same, calculates how much the actual means deviate from the assumption, and reports it as the F-test score. A larger score means there is a larger difference between the means.

**P-value**:  P-value tells how statistically significant is our calculated score value

####Note: If our price variable is strongly correlated with the variable we are analyzing, expect ANOVA to return a sizeable F-test score and a small p-value. Also,  the ANOVA algorithm averages the data automatically, no need to take the average before.

```
grouped_test2=df_gptest[['drive-wheels','price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```

```
grouped_test2.get_group('4wd')['price']
```

```
#ANOVA Algorithm 
from scipy import stats
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'], grouped_test2.get_group('4wd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

This is a great result, with a large F test score showing a strong correlation and a P value of almost 0 implying almost certain statistical significance. But does this mean all three tested groups are all this highly correlated? Let's see the results for other groups and variable-

```
#separately fwd-rwd

f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val )
```

```
#4wd-rwd
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('rwd')['price'])  
   
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```
#4wd-fwd
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('fwd')['price'])  
 
print("ANOVA results: F=", f_val, ", P =", p_val)   
```
